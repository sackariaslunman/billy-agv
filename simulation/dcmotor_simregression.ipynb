```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read raw data
df_raw = pd.read_excel("./datasets/sim/dcmotor_simdata.xls")
df_raw = df_raw.rename(columns = {
    "Var1": "t",
    "Var2": "RPM",
    "Var3": "Torque",
    "Var4": "Va",
    "Var5": "Ia",
})
df_raw["Wm"] = df_raw["RPM"] / 60 * 2 * np.pi
df_raw = df_raw.drop(columns=["Torque", "RPM", "Var6"])
df_raw
```

```python
from dcmotor import five_point_stencil

# Get derivatives
df = pd.DataFrame(df_raw)

time_step = df["t"].diff().iloc[1]

df["dIa"] = five_point_stencil(df["Ia"].array, time_step)
df["dWm"] = five_point_stencil(df["Wm"].array, time_step)

plt.plot(df["t"], df["Va"], c="blue")
plt.plot(df["t"], df["Ia"], c="orange")
plt.plot(df["t"], df["dIa"]/10, c="yellow")
plt.plot(df["t"], df["Wm"], c="red")
plt.plot(df["t"], df["dWm"], c="pink")
plt.show()
df
```

```python
# Linear regression

# Circuit Equation
# Va = i*Ra + L*di/dt + Ke*w
# Tm = Ke*i = TL + Jm*dw/dt + Bm*w
Ra = 0.091
Ke = 18.803 / 1000
TL = 0

t = np.array(df["t"].array)

Va = np.array(df["Va"].array)

Ia = np.array(df["Ia"].array)
dIa = np.array(df["dIa"].array)

Wm = np.array(df["Wm"].array)
dWm = np.array(df["dWm"].array)
```

```python
Xy = np.zeros((3,1))
XX_T = np.zeros((3,3))

# Va = i*Ra + L*di/dt + Ke*w 
# Kt*i - TL = 0 + Jm*dw/dt + Bm*w

for k in range(len(t)):
    y = np.vstack([Va[k]])
    X = np.array([
        [Ia[k], dIa[k], Wm[k]],
    ])
    Xy += X.T @ y
    XX_T += X.T @ X 

theta = np.linalg.inv(XX_T) @ Xy
Ra_pred, La_pred, Ke_pred = theta.T[0][:3]

print("Ra", Ra_pred)
print("La", La_pred)
print("Ke", Ke_pred)
```

```python
Xy = np.zeros((2,1))
XX_T = np.zeros((2,2))

# Kt*i = Jm*dw/dt + Bm*w + TL

for k in range(len(t)):
    y = np.vstack([Ke_pred * Ia[k]])
    X = np.array([
        [dWm[k], Wm[k]],
    ])
    Xy += X.T @ y
    XX_T += X.T @ X 

theta = np.linalg.inv(XX_T) @ Xy
Jm_pred, Bm_pred = theta.T[0][:2]

print("Jm", Jm_pred)
print("Bm", Bm_pred)
```

```python
# Circuit Equation  
# Va = i*Ra + L*di/dt + Ke*w
# Tm = Kt*i = TL + Jm*dw/dt + Bm*w
# TL = Kt*i - Jm*dw/dt - Bm*w

Va_pred = Ia*Ra + La_pred*dIa + Ke_pred*Wm
Tm = Ke_pred*Ia
Tm_pred = TL + Jm_pred*dWm + Bm_pred*Wm

plt.plot(t, Va, c="lightblue")
plt.plot(t, Va_pred, c="blue")
plt.show()

plt.plot(t, Tm, c="yellow")
plt.plot(t, Tm_pred, c="orange")
plt.show()
```
